## Teachable Machine Model Testing

This notebook tests the implementation of the rock-paper-scissors classifier using different images from each class.

### 1. Import Libraries

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### 2. Load Model and Labels

```python
model = tf.keras.models.load_model('keras_model.h5')

def load_labels(label_file):
    with open(label_file, 'r') as f:
        return f.read().splitlines()

labels = load_labels('labels.txt')
```

### 3. Function to Preprocess Image

```python
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (224, 224))
    img = np.array(img, dtype=np.float32) / 255.0
    img = np.expand_dims(img, axis=0)
    return img
```

### 4. Test the Model

```python
image_path = 'p.jpg'
img = preprocess_image(image_path)
predictions = model.predict(img)
class_idx = np.argmax(predictions)
confidence = predictions[0][class_idx]

print(f"Class: {labels[class_idx]}")
print(f"Confidence Score: {confidence:.4f}")

plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
plt.title(f"Prediction: {labels[class_idx]} ({confidence:.4f})")
plt.show()
```
