# Multi-Layer Perceptron Implementation and Testing

### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from multilayer_percetron import MultiLayerPerceptron
```

### Step Function Implementation

```python
def step(x):
    return (x > 0).astype(int)

x = np.arange(-10.0, 10.0, 0.1)
y = step(x)

plt.plot(x, y)
plt.title('Step Function')
plt.show()
```

### Sigmoid Function Plot



```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

y = sigmoid(x)
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.show()
```

### Testing the Multi-Layer Perceptron


```python
mlp = MultiLayerPerceptron(activation='sigmoid')
mlp.init_network()
```

### Testing with different inputs

```python
inputs = np.array([[0.5, 0.1], [1.0, -1.0], [-0.5, 0.7]])
outputs = [mlp.forward(i) for i in inputs]

for i, output in enumerate(outputs):
    print(f"Input: {inputs[i]}, Output: {output}")
```

### Visualizing Decision Boundaries

```python
from sklearn.datasets import make_moons
```


### Generating 2D dataset

```python
X, y = make_moons(n_samples=100, noise=0.1, random_state=42)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)
plt.title("2D Dataset (Moons)")
plt.show()
```

### Train MLP

```python
mlp = MultiLayerPerceptron(activation='sigmoid')
mlp.init_network()
```

### Define a grid for visualization

```python
xx, yy = np.meshgrid(np.linspace(-1.5, 2.5, 100), np.linspace(-1.0, 1.5, 100))
grid = np.c_[xx.ravel(), yy.ravel()]
preds = np.array([mlp.forward(point) for point in grid])
preds = preds.reshape(xx.shape)

plt.contourf(xx, yy, preds, alpha=0.8, cmap=plt.cm.Spectral)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='black', cmap=plt.cm.Spectral)
plt.title("Decision Boundary")
plt.show()
```
