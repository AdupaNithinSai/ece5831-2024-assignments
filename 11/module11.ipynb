#### Importing necessary modules

```python
from imdb import Imdb
from reuters import Reuters
from boston_housing import BostonHousing
```

#### IMDB Binary Classification

```python
imdb_model = Imdb()
imdb_model.prepare_data()
imdb_model.build_model()
imdb_model.train(epochs=20)
imdb_model.plot_metrics()
imdb_loss, imdb_accuracy = imdb_model.evaluate()
print(f"IMDB Test Loss: {imdb_loss}, Test Accuracy: {imdb_accuracy}")
```

#### Reuters Multiclass Classification

```python
reuters_model = Reuters()
reuters_model.build_model()
reuters_model.train(epochs=20)
reuters_model.plot_metrics()
reuters_loss, reuters_accuracy = reuters_model.evaluate()
print(f"Reuters Test Loss: {reuters_loss}, Test Accuracy: {reuters_accuracy}")
```

#### Boston Housing Regression


```python
boston_model = BostonHousing()
boston_model.prepare_data()
boston_model.build_model()
boston_model.train(epochs=20)
boston_model.plot_metrics()
boston_loss, boston_mae = boston_model.evaluate()
print(f"Boston Housing Test Loss: {boston_loss}, Test MAE: {boston_mae}")
```
